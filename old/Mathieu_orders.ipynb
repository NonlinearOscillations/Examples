```julia
using HarmonicBalance
```

```julia
@variables ω, ω0, F, t, α, γ,λ, x(t)
diff_eq = DifferentialEquation(d(x,t,2) + ω0^2 * (1 + λ * cos(ω*t) ) * x + α*x^3 + γ*d(x,t), x);
add_harmonic!(diff_eq, x, [0, ω, 2ω]);
harmonic_eq = get_harmonic_equations(diff_eq)
```

```julia
varied = ω => LinRange(0.93, 1.01, 100);
fixed = (α => 1., ω0 => 1, λ => 0.5, γ => 0.01);
```

```julia
result = get_steady_states(harmonic_eq, varied, fixed);
```

```julia
plot(result, x="ω", y="sqrt(u1^2 + v1^2)")
```

```julia
plot(result, x="ω", y="a1")
```
